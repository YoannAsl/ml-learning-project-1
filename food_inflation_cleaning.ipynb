```python
import pandas as pd
```

```python
df = pd.read_csv('data/Food price inflation - FAO_CP_23014.csv')
df
```

```python
column_to_check = 'TIME_PERIOD'
min_year = 2015
condition_to_keep = df[column_to_check].str[:4].astype(int) >= min_year
df_result = df[condition_to_keep].reset_index(drop=True)
df = df_result
df
```

```python
column_to_check = 'REF_AREA_LABEL'
condition_to_keep = ~df[column_to_check].isin(["South Asia","World","North America"])
df_result = df[condition_to_keep]
df = df_result
df
```

```python
uniques = df.nunique()
cols_to_drop = uniques[uniques <= 1].index
df_filtered = df.drop(columns=cols_to_drop)
df = df_filtered
df.rename(columns={'OBS_VALUE': 'Food price inflation index'}, inplace=True)
df
```

```python
df.rename(columns={'REF_AREA': 'Country Code'}, inplace=True)
df.rename(columns={'REF_AREA_LABEL': 'Country Name'}, inplace=True)
df.rename(columns={'TIME_PERIOD': 'Year'}, inplace=True)
df
```
